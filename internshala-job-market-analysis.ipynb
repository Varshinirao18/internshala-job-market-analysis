```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0"
}

job_title_list = []
company_list = []
location_list = []
salary_list = []
experience_list = []
posted_list = []
job_url_list = []

for page in range(1,15):   
    if page == 1:
        url = "https://internshala.com/jobs/"
    else:
        url = f"https://internshala.com/jobs/page-{page}/"
    
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")
    print(f"Scraped page {page}")
    job_cards = soup.find_all("div", class_="internship_meta")
     for card in job_cards:
        title_tag = card.find("a", class_="job-title-href")
        job_title_list.append(title_tag.get_text(strip=True) if title_tag else "")
        job_url_list.append("https://internshala.com" + title_tag.get("href") if title_tag else "")
        
        company = card.find("div", class_="heading_6 company_name")
        if company:
            company_text = company.get_text(strip=True)
            company_text = company_text.replace("Actively hiring", "").strip()
            company_list.append(company_text)
        else:
            company_list.append("")
        loc = card.find("p", class_="row-1-item locations")
        location_list.append(loc.get_text(strip=True) if loc else "")
         row_items = card.find_all("div", class_="row-1-item")
        if len(row_items) > 0:
            sal_span = row_items[0].find("span", class_="desktop")
            salary_list.append(sal_span.get_text(strip=True) if sal_span else "")
        else:
            salary_list.append("")

        experience_list.append(row_items[1].get_text(strip=True) if len(row_items) > 1 else "")
        
        posted = card.find("div", class_="color-labels")
        if posted:
            posted_list.append(posted.get_text(strip=True))
        else:
            posted_list.append("")
```

```python
df = pd.DataFrame({"Job Title": job_title_list,"Company": company_list,"Location": location_list,"Salary": salary_list,"Experience": experience_list,"Posted Date": posted_list,"Job URL": job_url_list})
```

# HYPOTHESIS TESTING

### 1.We want to test whether more than 50% of job postings are suitable for fresh graduates (0 years experience).

```python
from statsmodels.stats.proportion import proportions_ztest
success = (df2["Experience"] == 0).sum()
n = len(df2)
z_stat, p_value = proportions_ztest(success, n, value=0.5, alternative='larger')
p_value

```

+ Since p-value < 0.05, we reject H and conclude that most jobs are fresher-friendly.

## 2.We test whether the average salary offered to freshers is greater than 5 LPA.

```python
from scipy.stats import ttest_1samp
fresh_salary = df2[df2["Experience"] == 0]["Salary_num"]
t_stat, p_value = ttest_1samp(fresh_salary, 5, alternative='greater')
p_value
```

+ If p-value < 0.05, average fresher salary is significantly greater than 5 LPA.

## 3.We test whether average salaries differ between Bangalore and other cities.

```python
from scipy.stats import ttest_ind
blr = df2[df2["Location"].str.contains("Bangalore", case=False)]["Salary_num"]
others = df2[~df2["Location"].str.contains("Bangalore", case=False)]["Salary_num"]
t_stat, p_value = ttest_ind(blr, others, equal_var=False)
p_value
```

+ A p-value < 0.05 indicates significant salary differences across locations.

### 4.We want to test whether the average salary for experienced candidates (1 year) is significantly higher than that of freshers (0 years).

```python
from scipy.stats import ttest_ind
freshers = df2[df2["Experience"] == 0]["Salary_num"]
experienced = df2[df2["Experience"] >= 1]["Salary_num"]
t_stat, p_value = ttest_ind(experienced, freshers, equal_var=False, alternative="greater")
p_value
```

```python
exp_city_salary = df.groupby(["City", "Experience"])["Salary_num"].mean().reset_index()

top_cities = exp_city_salary["City"].value_counts().head(5).index
filtered = exp_city_salary[exp_city_salary["City"].isin(top_cities)]

pivot_table = filtered.pivot(index="Experience", columns="City", values="Salary_num")

pivot_table.plot(kind="bar", figsize=(10,5))
plt.title("Salary by Experience and City")
plt.xlabel("Experience (Years)")
plt.ylabel("Average Salary ( indicator)")
plt.xticks(rotation=0)
plt.show()
```

```python
role_exp_salary = df.groupby(["Job Title", "Experience"])["Salary_num"].mean().reset_index()

top_roles = role_exp_salary["Job Title"].value_counts().head(5).index
filtered_roles = role_exp_salary[role_exp_salary["Job Title"].isin(top_roles)]

for role in filtered_roles["Job Title"].unique():
    subset = filtered_roles[filtered_roles["Job Title"] == role]
    plt.plot(subset["Experience"], subset["Salary_num"], marker="o", label=role)

plt.title("Salary vs Experience for Top Job Roles")
plt.xlabel("Experience (Years)")
plt.ylabel("Average Salary ( indicator)")
plt.legend()
plt.show()
```
